**Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
import matplotlib.patches as patches
```

**Unstructured Triangular Grid**

```python
grid_point = pd.read_csv('grid_point.csv', header=None)
grid_triangles = pd.read_csv('grid_triangle.csv', header=None)
```

```python
grid_triangles = grid_triangles - 1
```

```python
triang = mtri.Triangulation(grid_point[0], grid_point[1], grid_triangles)
```

**Single Plot**

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.triplot(triang, color='black', linewidth=0.5)

plt.show()
```

**Create Multiple Axes**

```python
fig = plt.figure(figsize=(11,8))

# create the axis
ax1 = plt.subplot2grid(shape=(4,3), loc=(0,0), rowspan=4, colspan=2)

ax2 = plt.subplot2grid(shape=(4,3), loc=(0,2), rowspan=2)

ax3 = plt.subplot2grid(shape=(4,3), loc=(2,2), rowspan=2)

fig.tight_layout()

plt.show()
```

**Plot the Data**

```python
fig = plt.figure(figsize=(9,7), dpi=150)

# create the axis
ax1 = plt.subplot2grid(shape=(4,3), loc=(0,0), rowspan=4, colspan=2)
ax2 = plt.subplot2grid(shape=(4,3), loc=(0,2), rowspan=2)
ax3 = plt.subplot2grid(shape=(4,3), loc=(2,2), rowspan=2)

# plot the data: ax1 full model grid
ax1.triplot(triang, color='black', linewidth=0.5)

# plot the data: ax2 narrow strait
ax2.triplot(triang, color='black', linewidth=0.5)
ax2.set_xlim(210000, 230000) # 20,000 m
ax2.set_ylim(9085000, 9115000) # 30,000 m

# plot the data: ax3 Kuta Beach
ax3.triplot(triang, color='black', linewidth=0.5)
ax3.set_xlim(280000, 300000) # 20,000 m
ax3.set_ylim(9020000, 9050000) # 30,000 m

fig.tight_layout()

plt.show()
```

**Customization and Saving the Result**

```python
fig = plt.figure(figsize=(9,7), dpi=150)

# create the axis
ax1 = plt.subplot2grid(shape=(4,3), loc=(0,0), rowspan=4, colspan=2)
ax2 = plt.subplot2grid(shape=(4,3), loc=(0,2), rowspan=2)
ax3 = plt.subplot2grid(shape=(4,3), loc=(2,2), rowspan=2)

# plot the data: ax1 full model grid
ax1.triplot(triang, color='black', linewidth=0.5)
ax1.set_xlim(201000, 305000)
ax1.set_ylim(8995900, 9130000)
ax1.set_yticks(ticks=[9000000, 9020000, 9040000, 9060000, 9080000, 9100000, 9120000])
ax1.set_xticks(ticks=[210000, 230000, 250000, 270000, 290000])
ax1.set_yticklabels(ax1.get_yticks(), rotation=90, fontsize=10, va='center')
ax1.set_xticklabels(ax1.get_xticks(), fontsize=10, ha='center')

# plot the data: ax2 narrow strait
ax2.triplot(triang, color='black', linewidth=0.5)
ax2.set_xlim(210000, 230000) # 20,000 m
ax2.set_ylim(9085000, 9115000) # 30,000 m
ax2.set_yticks(ticks=[9090000, 9100000, 9110000])
ax2.set_xticks(ticks=[215000, 225000])
ax2.set_yticklabels(ax2.get_yticks(), rotation=90, fontsize=10, va='center')
ax2.set_xticklabels(ax2.get_xticks(), fontsize=10, ha='center')

# plot the data: ax3 Kuta Beach
ax3.triplot(triang, color='black', linewidth=0.5)
ax3.set_xlim(280000, 300000) # 20,000 m
ax3.set_ylim(9020000, 9050000) # 30,000 m
ax3.set_yticks(ticks=[9025000, 9035000, 9045000])
ax3.set_xticks(ticks=[285000, 295000])
ax3.set_yticklabels(ax3.get_yticks(), rotation=90, fontsize=10, va='center')
ax3.set_xticklabels(ax3.get_xticks(), fontsize=10, ha='center')

# add rectangle
ax1.add_patch(patches.Rectangle(xy=(210000, 9085000), width=20000, height=30000, color='red', fill=False))
ax1.add_patch(patches.Rectangle(xy=(280000, 9020000), width=20000, height=30000, color='red', fill=False))

# add x and y common label
fig.text(x=0.5, y=0, s='Eastings (UTM)', ha='center', fontsize=10, fontweight='bold') 
fig.text(x=0, y=0.5, s='Northings (UTM)', va='center', fontsize=10, fontweight='bold', rotation='vertical')

# add annotation text
ax1.text(205000, 9125000, '(a)', fontsize=10, fontweight='bold')
ax2.text(211000, 9112500, '(b)', fontsize=10, fontweight='bold')
ax3.text(297000, 9048000, '(c)', fontsize=10, fontweight='bold')

fig.tight_layout()

fig.savefig('Multiple Plots with different size axis.png', dpi=150, bbox_inches='tight')

plt.show()
```
