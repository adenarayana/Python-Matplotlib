# **3. Plot the Velocity Vector (default)**

# **4. Unstructured Triangular Grid**

```python
grid_point = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/003_VelocityVectorPlot/grid_point.csv',header=None)
grid_triangle = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/003_VelocityVectorPlot/grid_triangle.csv', header=None)
```

```python
grid_triangle = grid_triangle - 1
```

```python
triang = mtri.Triangulation(grid_point[0], grid_point[1], triangles=grid_triangle)
```

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.triplot(triang, color='black', linewidth=0.5)
ax.quiver(center_x, center_y, u, v)

plt.show()
```

# **5. Plot Grid and Velocity Vector**

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.triplot(triang, color='black', linewidth=0.5, markersize=5, markerfacecolor='black', markeredgecolor='black')
ax.quiver(center_x, center_y, u, v)

ax.set_xlim(224000,229000)
ax.set_ylim(9025000,9030000)

ax.set_ylabel('y axis', fontweight='bold', fontsize=12)
ax.set_xlabel('x axis', fontweight='bold', fontsize=12)

plt.show()
```

# **6. Customized the Arrow**

## Arrow Length: scale

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.triplot(triang, color='black', linewidth=0.5, markersize=5, markerfacecolor='black', markeredgecolor='black')
ax.quiver(center_x, center_y, u, v, scale=0.1)

ax.set_xlim(224000,229000)
ax.set_ylim(9025000,9030000)

ax.set_ylabel('y axis', fontweight='bold', fontsize=12)
ax.set_xlabel('x axis', fontweight='bold', fontsize=12)

plt.show()
```

## Arrow shape: width, headwidth, headlength, headaxislength

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.triplot(triang, color='black', linewidth=0.5, markersize=5, markerfacecolor='black', markeredgecolor='black')

ax.quiver(center_x, center_y, u, v, 
          scale=0.1, 
          width=0.005, 
          headwidth=7.5, 
          headlength=10, 
          headaxislength=5)

ax.set_xlim(224000,229000)
ax.set_ylim(9025000,9030000)

ax.set_ylabel('y axis', fontweight='bold', fontsize=12)
ax.set_xlabel('x axis', fontweight='bold', fontsize=12)

plt.show()
```

## x and y ticks label

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.triplot(triang, color='black', linewidth=0.5, markersize=5, markerfacecolor='black', markeredgecolor='black')

ax.quiver(center_x, center_y, u, v, 
          scale=0.1, 
          width=0.005, 
          headwidth=7.5, 
          headlength=10, 
          headaxislength=5)

ax.set_xlim(224000,229000)
ax.set_ylim(9025000,9030000)

ax.set_ylabel('y axis', fontweight='bold', fontsize=12)
ax.set_xlabel('x axis', fontweight='bold', fontsize=12)

ax.set_yticks(ticks=[9025500, 9027500, 9029500])
ax.set_xticks(ticks=[224500, 226500, 228500])

ax.set_yticklabels(ax.get_yticks(), rotation=90, fontsize=12, va='center')
ax.set_xticklabels(ax.get_xticks(), fontsize=12, ha='center')

plt.show()
```

# **7. Adding an Arrow Legend**

```python
# data for legend
legend_x = 224500
legend_y = 9029500

legend_u = 0.01 # in m/s
legend_v = 0

# plot
fig, ax = plt.subplots(figsize=(8,8))

ax.triplot(triang, color='black', linewidth=0.5, markersize=5, markerfacecolor='black', markeredgecolor='black')

ax.quiver(center_x, center_y, u, v, 
          scale=0.1, 
          width=0.005, 
          headwidth=7.5, 
          headlength=10, 
          headaxislength=5)

ax.quiver(legend_x, legend_y, legend_u, legend_v,
          scale=0.1, 
          width=0.005, 
          headwidth=7.5, 
          headlength=10, 
          headaxislength=5) 

ax.set_xlim(224000,229000)
ax.set_ylim(9025000,9030000)

ax.set_ylabel('y axis', fontweight='bold', fontsize=12)
ax.set_xlabel('x axis', fontweight='bold', fontsize=12)

ax.set_yticks(ticks=[9025500, 9027500, 9029500])
ax.set_xticks(ticks=[224500, 226500, 228500])

ax.set_yticklabels(ax.get_yticks(), rotation=90, fontsize=12, va='center')
ax.set_xticklabels(ax.get_xticks(), fontsize=12, ha='center')

ax.text(x=224500, y=9029610, s='0.01 m/s', fontsize=12)

plt.show()
```

# **8. Get the Triangle Center Location**

```python
grid_point = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/003_VelocityVectorPlot/grid_point.csv',header=None)
grid_triangle = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/003_VelocityVectorPlot/grid_triangle.csv', header=None)
```

```python
grid_triangle = grid_triangle - 1
```

```python
center_x = []
center_y = []

for i in range(0, len(grid_triangle)):
  center_x.append((grid_point[0][grid_triangle[0][i]] + grid_point[0][grid_triangle[1][i]] + grid_point[0][grid_triangle[2][i]])/3)
  center_y.append((grid_point[1][grid_triangle[0][i]] + grid_point[1][grid_triangle[1][i]] + grid_point[1][grid_triangle[2][i]])/3)

center_x = np.asarray(center_x)
center_y = np.asarray(center_y)
```
