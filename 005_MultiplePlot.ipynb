**Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
```

**Load the Data**

```python
# unstructured grid
grid_point = pd.read_csv('grid_point.csv', header=None)
grid_triangle = pd.read_csv('grid_triangle.csv', header=None)
```

```python
# water surface elevation 
zeta_01 = np.array(pd.read_csv('zeta_303.csv', header=None)).reshape(-1) # flood tide
zeta_02 = np.array(pd.read_csv('zeta_307.csv', header=None)).reshape(-1) # high tide
zeta_03 = np.array(pd.read_csv('zeta_310.csv', header=None)).reshape(-1) # ebb tide
zeta_04 = np.array(pd.read_csv('zeta_313.csv', header=None)).reshape(-1) # low tide
```

**Create Unstructured Triangular Grid**

```python
grid_triangle = grid_triangle - 1
```

```python
unstructured_grid = mtri.Triangulation(x=grid_point[0],
                                      y=grid_point[1], 
                                      triangles=grid_triangle)
```

**Single Contour Fill Plot**

```python
fig, ax = plt.subplots(figsize=(5,7))

ax.tricontourf(unstructured_grid, zeta_04, cmap='jet')

plt.show()
```

**Multiple Plot: Basic**

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(7,9))

plt.show()
```

**Customize the x and y ticks labels**

```python
color_levels = np.linspace(start=-0.7797577381134033, stop=0.7335739731788635, num=100)
color_map = 'jet'
```

```python
# creating figure and axes object
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,9), sharey=True, sharex=True)

# plot the data
ax[0,0].tricontourf(unstructured_grid, zeta_01, cmap=color_map, levels=color_levels)
ax[0,1].tricontourf(unstructured_grid, zeta_02, cmap=color_map, levels=color_levels)
ax[1,0].tricontourf(unstructured_grid, zeta_03, cmap=color_map, levels=color_levels)
ax[1,1].tricontourf(unstructured_grid, zeta_04, cmap=color_map, levels=color_levels)

# set the limit of x and y axes
ax[0,0].set_xlim(xlim_min, xlim_max)
ax[0,0].set_ylim(ylim_min, ylim_max)

# customize the x and y ticks labels
ax[0,0].set_xticks(ticks=[210000, 250000, 290000])
ax[0,0].set_yticks(ticks=[9020000, 9060000, 9100000])

# rotate the y axis
ax[0,0].set_yticklabels(ax[0,0].get_yticks(), rotation=90, va='center')
ax[1,0].set_yticklabels(ax[1,0].get_yticks(), rotation=90, va='center')

# add the common x and y labels
fig.text(x=0.5, y=0, s='Longitude (UTM)', ha='center', fontweight='bold') 
fig.text(x=0, y=0.5, s='Latitude (UTM)', va='center', fontweight='bold', rotation='vertical')

fig.tight_layout()

plt.show()
```

**Add a Colorbar**

```python
# creating figure and axes object
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,9), sharey=True, sharex=True)

# plot the data
color = ax[0,0].tricontourf(unstructured_grid, zeta_01, cmap=color_map, levels=color_levels)
ax[0,1].tricontourf(unstructured_grid, zeta_02, cmap=color_map, levels=color_levels)
ax[1,0].tricontourf(unstructured_grid, zeta_03, cmap=color_map, levels=color_levels)
ax[1,1].tricontourf(unstructured_grid, zeta_04, cmap=color_map, levels=color_levels)

# set the limit of x and y axes
ax[0,0].set_xlim(xlim_min, xlim_max)
ax[0,0].set_ylim(ylim_min, ylim_max)

# customize the x and y ticks labels
ax[0,0].set_xticks(ticks=[210000, 250000, 290000])
ax[0,0].set_yticks(ticks=[9020000, 9060000, 9100000])

# rotate the y axis
ax[0,0].set_yticklabels(ax[0,0].get_yticks(), rotation=90, va='center')
ax[1,0].set_yticklabels(ax[1,0].get_yticks(), rotation=90, va='center')

# add the common x and y labels
fig.text(x=0.5, y=0, s='Longitude (UTM)', ha='center', fontweight='bold') 
fig.text(x=0, y=0.5, s='Latitude (UTM)', va='center', fontweight='bold', rotation='vertical')

fig.tight_layout()

# adding colorbar
cax = fig.add_axes([1, 0.05, 0.04, 0.93])
cbar = fig.colorbar(color, cax=cax, orientation='vertical')

# customize the colorbar
cbar.ax.set_ylabel('Water Surface Elevation (m)', fontweight='bold')
cbar.ax.set_yticks(ticks=[-0.70, -0.35, 0.00, 0.35, 0.7])
cbar.ax.set_yticklabels([-0.70, -0.35, 0.00, 0.35, 0.7], rotation='vertical', va='center')

plt.show()
```
