# **Import the Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.tri as mtri
import matplotlib.pyplot as plt
```

# **Load the Data: Grid and Triangle**

```python
grid_point = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/002_TricontourPlot/grid_point.csv', header=None)
grid_triangle = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/002_TricontourPlot/grid_triangle.csv', header=None)
grid_depth = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/002_TricontourPlot/grid_depth.csv', header=None)
```

```python
x = grid_point[0].values
y = grid_point[1].values
```

```python
x.shape
```

```python
grid_depth.shape
```

```python
grid_depth = np.array(grid_depth).reshape(-1)
```

```python
grid_depth.shape
```

# **Combine the Point and Triangle: Unstructured Triangular Grid**

```python
triang = mtri.Triangulation(x, y, triangles=grid_triangle-1)
```

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.triplot(triang, color='black', linewidth=0.5)

plt.show()
```

# **Draw the Tricontour Plot: Default**

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.tricontourf(triang, grid_depth)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.tricontourf(x, y, grid_depth)

plt.show()
```

# **Draw the Tricontour Plot: Customized**

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.tricontourf(triang, grid_depth, cmap='jet')

ax.set_ylabel('Latitude (UTM)', fontweight='bold', fontsize=12)
ax.set_xlabel('Longitude (UTM)', fontweight='bold', fontsize=12)

ax.set_yticks(ticks=[9020000, 9040000, 9060000, 9080000, 9100000, 9120000])
ax.set_xticks(ticks=[210000, 230000, 250000, 270000, 290000])

ax.set_yticklabels(ax.get_yticks(), rotation=90, fontsize=12, va='center')
ax.set_xticklabels(ax.get_xticks(), fontsize=12, ha='center')

ax.set_xlim(201000, 305000)
ax.set_ylim(9010000, 9130000)

plt.show()
```

# **Adding Colorbar**

```python
fig, ax = plt.subplots(figsize=(8,12))

# draw the contour fill
im = ax.tricontourf(triang, grid_depth, cmap='jet', levels=50)

# adding the x and y axis labels
ax.set_ylabel('Latitude (UTM)', fontweight='bold', fontsize=12)
ax.set_xlabel('Longitude (UTM)', fontweight='bold', fontsize=12)

# set the x and y ticks
ax.set_yticks(ticks=[9020000, 9040000, 9060000, 9080000, 9100000, 9120000])
ax.set_xticks(ticks=[210000, 230000, 250000, 270000, 290000])

# rotate the yticks label and change the fontsize
ax.set_yticklabels(ax.get_yticks(), rotation=90, fontsize=12, va='center')
ax.set_xticklabels(ax.get_xticks(), fontsize=12, ha='center')

# set the plot axis limit
ax.set_xlim(201000, 305000)
ax.set_ylim(9010000, 9130000)

# colorbar position
cax = fig.add_axes([0.92, 0.15, 0.04, 0.7])

# adding colorbar and customize the colorbar
cbar = fig.colorbar(im, cax=cax, orientation='vertical')
cbar.ax.set_yticklabels(cbar.ax.get_yticklabels(), rotation='vertical', fontsize=12, va='center')
cbar.ax.set_ylabel('Color Map', fontweight='bold', fontsize=12)

# save the figure
fig.savefig('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/002_TricontourPlot/contourfill.png',dpi=150, bbox_inches='tight')

plt.show()
```
