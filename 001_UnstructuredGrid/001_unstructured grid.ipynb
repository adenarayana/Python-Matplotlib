# **Import the Libraries**

```python
import pandas as pd
import matplotlib.tri as mtri
import matplotlib.pyplot as plt
```

# **Load the Data: Grid and Triangle**

```python
grid_point = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/grid_point.csv', header=None)
grid_triangle = pd.read_csv('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/grid_triangle.csv', header=None)
```

```python
grid_point
```

```python
grid_triangle-1
```

# **Combine the Point and Triangle**

```python
triang = mtri.Triangulation(x=grid_point[0], 
                            y=grid_point[1], 
                            triangles=grid_triangle-1)
```

# **Draw an Unstructured Triangular Grid: Default**

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.triplot(triang)

plt.show()
```

# **Draw an Unstructured Triangular Grid: Customized**

```python
fig, ax = plt.subplots(figsize=(8,12))

ax.triplot(triang, color='black', linewidth=0.5)

ax.set_ylabel('Latitude (UTM)', fontweight='bold', fontsize=12)
ax.set_xlabel('Longitude (UTM)', fontweight='bold', fontsize=12)

ax.set_yticks(ticks=[9020000, 9040000, 9060000, 9080000, 9100000, 9120000])
ax.set_xticks(ticks=[210000, 230000, 250000, 270000, 290000])

ax.set_yticklabels(ax.get_yticks(), rotation=90, fontsize=12, va='center')
ax.set_xticklabels(ax.get_xticks(), fontsize=12, ha='center')

fig.savefig('/content/drive/MyDrive/Youtube Tutorial/Matplotlib Tutorial/#1 plot unstructured grid/result.png', dpi=150, bbox_inches='tight')

plt.show()
```
